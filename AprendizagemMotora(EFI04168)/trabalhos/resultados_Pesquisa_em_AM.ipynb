```python
# Chamando minha biblioteca
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importar a planilha do excel
dados = pd.read_excel("/content/dados.xlsx")
```

```python
# Separando os dados por grupos
gv = dados.loc[dados['grupos'] == 'GV']
gc = dados.loc[dados['grupos']=='GC']
```

```python
#Obtendo a média de cada grupo e de cada momento
gv_media_pre = gv["pre-teste"].mean()

gc_media_pre = gc["pre-teste"].mean()

gv_media_pos = gv["pos-teste"].mean()

gc_media_pos = gc["pos-teste"].mean()
```

```python
plt.plot([gv_media_pre,gv_media_pos], '-or',  label='Grupo Variada')
plt.plot([gc_media_pre,gc_media_pos], '-og', label='Grupo Constante')

# Adicionando rótulo ao eixo y
plt.ylabel('Pontos no forehand')

# Adicionando uma legenda
plt.legend()

# Personalizando os rótulos do eixo x
plt.xticks([0, 1], ['pré-teste', 'pós-teste'])


# Adicionando uma legenda
plt.legend()

# Exibindo o gráfico
plt.show()
```
